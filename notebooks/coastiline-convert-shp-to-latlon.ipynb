```python
import fiona
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
# from shapely.geometry import shape

coordinates = []

with fiona.open(r'./../data/geo-data/ne_10m_coastline/ne_10m_coastline.shp', 'r') as features:
    print(f"The file has {len(features)} features.")
    for feature in features:
        coords = feature['geometry']['coordinates']
        coordinates = coordinates + coords
```

```python
df = pd.DataFrame(coordinates, columns=["lon", "lat"])

fig,ax = plt.subplots(figsize=(20, 9))
ax.plot(df.lon, df.lat, '.', ms=0.2)
ax.set_xlabel("lon")
ax.set_ylabel("lat")

ax.plot(-105.252644, 40.023222, 'o', c='r', ms=10)
```

```python
df.to_csv('./../data/geo-data/shorelines_lat_lon.csv', index=False)
```
