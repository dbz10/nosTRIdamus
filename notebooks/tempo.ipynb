```python
from bs4 import BeautifulSoup
import re
```

```python
with open('./../data/tempo/costa.html', 'r') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html)
```

```python
# let's target the general description of the event
description = soup.find('div', attrs={'class': "moduleWrap genericDesc eventDescription"}).get_text(' ')
# remove all redondant spaces, tabs, etc ...
description = re.sub("\s+", " ", description).strip()
description
```

```python
tokens = [t for t in description.split()]
tokens
```
