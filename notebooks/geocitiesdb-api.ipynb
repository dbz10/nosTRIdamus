```python
import requests
import pandas as pd
import json
# from config import Cfg as cfg
```

```python
df_races = pd.read_csv('./../data/geo-data/races_lat_lon.csv')
df_races.head()
```

```python
races_metropolitan_area = {}
    
for race in df_races.race:
    lat = df_races.loc[df_races.race == race, 'lat'].values[0]
    lng = df_races.loc[df_races.race == race, 'lon'].values[0]
    races_metropolitan_area[race] = {'lat': lat, 'lng': lng}
    
    # geocitiesdb API
    lat_str = f'+{lat}' if lat > 0 else f'{lat}'
    lng_str = f'+{lng}' if lng > 0 else f'{lng}'
    minPopulation = 300000
    radius = 100
    unit = 'MI'
    
    requestString = f'http://geodb-free-service.wirefreethought.com/v1/geo/locations/{lat_str}{lng_str}/nearbyCities?limit=10&offset=0&minPopulation={minPopulation}&radius={radius}&distanceUnit={unit}'

    r = requests.get(requestString)
    response = r.json()
    
    races_metropolitan_area[race]['totalCount'] = response['metadata']['totalCount']
    races_metropolitan_area[race]['data'] = response['data']

    # get city details
    for city in races_metropolitan_area[race]['data']:
        cityId = city['id']
        url_city = f'http://geodb-free-service.wirefreethought.com/v1/geo/cities/{cityId}'
        r = requests.get(url_city)
        city.update(r.json()['data'])

# save data
with open(f'./../data/geo-data/races-metropolitan-area.json', 'w') as f:
    f.write(json.dumps(races_metropolitan_area))
```
